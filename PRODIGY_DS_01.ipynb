```python
# import the libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#identifying categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns

# displaying/printing the categorical variables
print("Categorical variables: ")
for column in categorical_columns:
    print(column)
```

```python
# Display the percentage distribution for each categorical variable
for column in categorical_columns:
    print(f"{column.capitalize()} Distribution:")
    value_counts = df[column].value_counts(normalize=True) * 100  # Convert counts to percentages
    print(value_counts)
    print()
```

### visualisation 

```python
# creating a bargraph for each categorical variable 
```

```python
# gender distribution
plt.figure(figsize=(8,5))
gender_counts=df['gender'].value_counts()
gender_counts.plot(kind='bar',color=['pink','blue'])
plt.title("Gender distribution")
plt.xlabel("Gender")
plt.ylabel("counts")
plt.xticks(rotation=0)
plt.show()
gender_counts
```

```python
# ethnicity distribution
plt.figure(figsize=(8,5))
race_counts=df['race/ethnicity'].value_counts()
race_counts.plot(kind='bar',color=['red'])
plt.title("Ethnicity distribution")
plt.xlabel("Ethnicity")
plt.ylabel("counts")
plt.xticks(rotation=0)
plt.show()
race_counts
```

```python
#Parental level of education
plt.figure(figsize=(8,5))
ple_counts=df['parental level of education'].value_counts()
ple_counts.plot(kind='bar')
plt.title("Parental level of education distribution")
plt.xlabel("Parental level of education")
plt.ylabel("counts")
plt.xticks(rotation=90)
plt.show()
ple_counts
```

```python
# lunch distribution
plt.figure(figsize=(8,5))
lunch_counts=df['lunch'].value_counts()
lunch_counts.plot(kind='bar',color='green')
plt.title("lunch distribution")
plt.xlabel("lunch")
plt.ylabel("counts")
plt.xticks(rotation=0)
plt.show()
lunch_counts
```

```python
plt.figure(figsize=(8, 5))
test_prep_counts = df['test preparation course'].value_counts()
test_prep_counts.plot(kind='bar', color=['red', 'cyan'])
plt.title('Test Preparation Course Distribution')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
test_prep_counts
```

```python
# continuous variables
continuous_columns = df.select_dtypes(include=['int64']).columns

# displaying/printing the continuous variables
print("continuous variables: ")
for column in continuous_columns:
    print(column)
```

```python
# plotting histogram for continuous variables
for column in continuous_columns:
    plt.figure(figsize=(8, 5))
    plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {column.replace(" ", "/").capitalize()}')
    plt.xlabel(column.capitalize())
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
